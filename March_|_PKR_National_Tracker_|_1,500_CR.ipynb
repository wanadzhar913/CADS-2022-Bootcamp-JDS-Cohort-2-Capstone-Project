## 1.4 Adjust column data types and simple EDA

```python
live.phonenum = live.phonenum.astype('int64')
```

```python
# Select all columns except for `phonenum` and change them to the categorical datatype
columns_to_convert = [col for col in live.columns if col != 'phonenum']
live[columns_to_convert] = live[columns_to_convert].astype('category')
```

```python
# Clean responses
live = live.replace(dynamo)
```

```python
# see raw count breakdown of each variablee

category_col = []
for column in live.columns:
     if live[column].dtype == 'category' and len(live[column].unique()) <= 15:
        category_col.append(column)
        print(f"{column} : {live[column].unique()}")
        print(live[column].value_counts() )
        print("====================================")
```

```python
# Convert text in Q3 into titlecase format, where every first character in every word is upper case
live['3. Where do you get most of your updates about current issues from?'] = live['3. Where do you get most of your updates about current issues from?'].str.title()
```

## 1.6 Create `Region` column
based on `State` column

```python
# Initialize the Region column
live['Region'] = 'null'
```

```python
# Set Region that corresponds to State
live.loc[live['State'].isin(['Pahang', 'Terengganu', 'Kelantan']), 'Region'] = 'Northeast'

live.loc[live['State'].isin(['Perlis', 'Kedah', 'Pulau Pinang']), 'Region'] = 'Northwest'

live.loc[live['State'].isin(['Perak', 'Selangor', 'Kuala Lumpur', 'Putrajaya', 'Negeri Sembilan']), 'Region'] = 'Central'

live.loc[live['State'].isin(['Melaka', 'Johor']), 'Region'] = 'South'

live.loc[live['State'].isin(['Sabah', 'Sarawak', 'Labuan']), 'Region'] = 'East'
```
